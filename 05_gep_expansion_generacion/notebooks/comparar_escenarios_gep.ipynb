# Exploración de datos GEP

```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

base = Path('../actividades')
demanda = pd.read_csv(base/'datos_actividad_05_demanda.csv')
tecnologias = pd.read_csv(base/'datos_actividad_05_tecnologias.csv')
existentes = pd.read_csv(base/'datos_actividad_05_existentes.csv')

display(tecnologias)
display(existentes)
```

```python
ax = tecnologias.plot(kind='bar', x='tecnologia', y='capex_kusd_mw', legend=False)
ax.set_ylabel('CAPEX [kUSD/MW]')
ax.set_title('CAPEX de tecnologías candidatas')
plt.show()
```
