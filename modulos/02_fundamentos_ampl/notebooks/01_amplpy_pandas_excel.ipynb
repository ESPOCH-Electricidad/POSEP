# AMPL, pandas y Excel
Flujo Excel → pandas → .dat.


```python
from pathlib import Path
import pandas as pd
excel_path = Path('../excel/pintura_ampl.xlsx')
products = pd.read_excel(excel_path, sheet_name='products')
params = pd.read_excel(excel_path, sheet_name='parameters')
products
```

```python
dat_lines = []
dat_lines.append('set P := ' + ' '.join(products['PRODUCT'].astype(str)) + ';')
for col in ['price','rate','market']:
    dat_lines.append(f'param {col} :=')
    for _, row in products.iterrows():
        dat_lines.append(f"{row['PRODUCT']} {row[col]}")
    dat_lines.append(';')
hours = params.loc[params['parameter']=='Hours','value'].iloc[0]
dat_lines.append(f'param Hours := {hours};')
Path('pintura_generado.dat').write_text('\n'.join(dat_lines), encoding='utf-8')
print('\n'.join(dat_lines))
```
