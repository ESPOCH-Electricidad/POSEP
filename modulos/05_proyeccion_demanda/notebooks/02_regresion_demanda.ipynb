# Regresión de demanda
Notebook didáctico para el módulo de proyección de demanda.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar datos del módulo
df = pd.read_csv('../datos/demanda_historica.csv')
df.head()
```

```python
plt.figure(figsize=(7,4))
plt.plot(df['anio'], df['energia_gwh'], marker='o')
plt.xlabel('Año')
plt.ylabel('Energía [GWh]')
plt.title('Demanda histórica')
plt.grid(True)
plt.show()
```
