# Exploración de datos de operación

Este notebook lee los datos tabulados de la actividad 02 y genera gráficos básicos.

```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

base = Path('../actividades')
gen = pd.read_csv(base/'datos_actividad_02_generadores.csv')
dem = pd.read_csv(base/'datos_actividad_02_demanda.csv')
hidro = pd.read_csv(base/'datos_actividad_02_hidro.csv')

display(gen)
display(dem)
```

```python
ax = dem.plot(x='hora', y='demanda_mw', marker='o', legend=False)
ax.set_xlabel('Hora')
ax.set_ylabel('Demanda [MW]')
ax.set_title('Demanda horaria')
plt.show()
```
