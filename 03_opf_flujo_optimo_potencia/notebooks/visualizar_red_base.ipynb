# Visualización de red OPF

Lee buses, líneas y generadores de la actividad 03.

```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

base = Path('../actividades')
buses = pd.read_csv(base/'datos_actividad_03_buses.csv')
lineas = pd.read_csv(base/'datos_actividad_03_lineas.csv')
generadores = pd.read_csv(base/'datos_actividad_03_generadores.csv')

display(buses)
display(lineas)
display(generadores)
```

```python
ax = lineas.plot(kind='bar', x='linea', y='limite_mw', legend=False)
ax.set_ylabel('Límite [MW]')
ax.set_title('Capacidad térmica por línea')
plt.show()
```
