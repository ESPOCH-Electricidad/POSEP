# Comparación de corredores candidatos TNEP

```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

base = Path('../actividades')
corredores = pd.read_csv(base/'datos_actividad_04_corredores.csv')
display(corredores)
```

```python
ax = corredores.plot(kind='bar', x='corredor', y='costo_kusd', legend=False)
ax.set_ylabel('Costo [kUSD]')
ax.set_title('Costo de inversión por corredor')
plt.show()
```
