# Exploración del caso Garver GEP

Este notebook permite revisar los datos públicos del caso antes de ejecutar el modelo AMPL entregado por el docente.

Objetivos:

1. Leer tecnologías existentes y candidatas.
2. Graficar demanda por periodo.
3. Comparar costo variable, costo de inversión, crédito firme y emisiones.
4. Verificar capacidad firme inicial frente a demanda y margen de reserva.

```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

base = Path('..') / 'datos_csv'
tech = pd.read_csv(base / 'generacion_tecnologias.csv')
demand = pd.read_csv(base / 'demanda_periodos.csv')
constraints = pd.read_csv(base / 'restricciones_periodos.csv')
limits = pd.read_csv(base / 'limites_inversion.csv')

tech
```

```python
demand.plot(x='year', y='Pdem_MW', marker='o', legend=False)
plt.xlabel('Año')
plt.ylabel('Demanda pico [MW]')
plt.title('Demanda del caso Garver GEP')
plt.grid(True)
plt.show()
```

```python
tech[['gen','tipo','tecnologia','Pmax0_MW','costo_variable','costo_inversion','credito_firme','factor_emisiones']]
```

```python
ax = tech.set_index('gen')[['costo_variable','costo_inversion','credito_firme','factor_emisiones']].plot(kind='bar', figsize=(10,5))
plt.title('Comparación de parámetros por tecnología')
plt.xlabel('Tecnología')
plt.grid(axis='y')
plt.show()
```

```python
rm = constraints['reserve_margin'].iloc[0]
capacidad_firme_inicial = (tech['Pmax0_MW'] * tech['credito_firme']).sum()
for _, row in demand.iterrows():
    req = (1+rm)*row['Pdem_MW']
    print(f"Año {int(row['year'])}: capacidad firme inicial = {capacidad_firme_inicial:.1f} MW, requerimiento = {req:.1f} MW, déficit = {max(0, req-capacidad_firme_inicial):.1f} MW")
```
