# Ejemplo mínimo ejecutable: despacho económico lineal

Este notebook es un ejemplo didáctico pequeño. No reemplaza los modelos AMPL del curso. Sirve para visualizar cómo un problema de optimización selecciona generación al menor costo.

```python
import pandas as pd
import matplotlib.pyplot as plt

gens = pd.DataFrame({
    'gen':['G1','G2','G3'],
    'Pmin':[0,0,0],
    'Pmax':[100,80,60],
    'cost':[20,35,50]
})
D = 150

gens
```

```python
# Algoritmo greedy válido solo para este caso lineal simple sin mínimos técnicos ni red.
remaining = D
dispatch = []
for _, g in gens.sort_values('cost').iterrows():
    p = min(g['Pmax'], remaining)
    dispatch.append((g['gen'], p, g['cost'], p*g['cost']))
    remaining -= p
    if remaining <= 1e-9:
        break

result = pd.DataFrame(dispatch, columns=['gen','P','cost','total_cost'])
print('Demanda no cubierta:', remaining)
result
```

```python
result.plot(x='gen', y='P', kind='bar', legend=False)
plt.ylabel('MW')
plt.title('Despacho económico mínimo')
plt.grid(axis='y')
plt.show()
print('Costo total:', result['total_cost'].sum())
```

## Discusión

Este ejemplo muestra la intuición de mérito económico. En modelos reales, esta regla puede fallar si existen pérdidas, congestión, mínimos técnicos, rampas, costos de arranque o restricciones de reserva.
